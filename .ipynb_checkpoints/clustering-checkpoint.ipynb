```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Load Dataset
df_penguins = pd.read_csv("penguins.csv")
print(df_penguins)
```

```python
# Data Pra-Preprocessing
# Data Checking : check data nya lebih dulu apakah ada outlier ataupun yang gak sesuai
print(df_penguins.head())
```

```python
print(df_penguins.info())
```

```python
print(df_penguins.isnull().sum())
```

```python
# berdasarkan data checking terlalu banyan NaN, drop aja
df_penguins_ii = df_penguins.dropna()
print(df_penguins)
```

```python
# sekarang cari bagian outliernya, outlier = data yang lebih atau kurang dari jumlah normalnya
df_penguins_ii.boxplot()
plt.show()
```

```python
print(df_penguins_ii[df_penguins_ii['flipper_length_mm']>4000],df_penguins_ii[df_penguins_ii['flipper_length_mm']<0])
```

```python
penguins_clear = df_penguins_ii.drop([9,14])
penguins_clear.boxplot()
plt.show()
```

```python
print(penguins_clear)
```

```python
# mengubah data kategorikal (sex) ke numerik
df = pd.get_dummies(penguins_clear, dtype='uint8').drop('sex_.',axis=1)
df.head()
```

```python
# Data Pre-Processing 

scaler = StandardScaler()

X = scaler.fit_transform(df)

penguins_preprocessed = pd.DataFrame(data=X,columns=df.columns)
penguins_preprocessed.head(10)
```

```python
%matplotlib inLine

pca = PCA()
pca.fit(penguins_preprocessed)
exp_variance = pca.explained_variance_ratio_

# ----- #

fig,ax = plt.subplots()
ax.bar(range(pca.n_components_), exp_variance)
ax.set_xlabel('Principal Component #')
ax.axhline(y=0.1, linestyle='--')
```

```python
# dari plot di atas, telah didapati bahwa terdapat dua fitur yang lumayan kuat
# ------ #

n_components = 2

pca = PCA(n_components=n_components, random_state=42)

penguins_PCA = pca.fit_transform(penguins_preprocessed)
```

```python
# selanjutnya pada tahap ini, kita coba untuk menentukan berapa jumlah cluster yang baik untuk data

inertia = []

for k in range(1,10):
    kmeans= KMeans(n_clusters=k, random_state=42).fit(penguins_PCA)
    inertia.append(kmeans.inertia_)
plt.plot(range(1, 10), inertia, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method')
plt.show()
```

```python
# nah, di atas kan dah dapat tuh jumlah clustering yang baiknya berapa, selanjutnya tinggal nerapin algoritma kmeansnya

n_clusters = 4

kmeans = KMeans(n_clusters = n_clusters, random_state=42).fit(penguins_PCA)

plt.scatter(penguins_PCA[:,0], penguins_PCA[:,1], c=kmeans.labels_, cmap='viridis')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
plt.title(f'K-means Clustering (K={n_clusters})')
plt.show()
```

```python
# Data

penguins_clear['label']=kmeans.labels_
numeric_columns = ['culmen_length_mm', 'culmen_depth_mm', 'flipper_length_mm', 'label']
stat_penguins = penguins_clear[numeric_columns].groupby('label').mean()

stat_penguins
```
